# Example: how to use Cv2Workaround in Google Colab
* Upload the module file (**cv2workaround.py**) on your Google Drive
* The following example uses **Utilities4Colab** folder on Google Drive
* **test.jpg** is loaded from the top directory of Google Drive

```python
# Mount Google Drive 
from google.colab import drive
drive.mount('/content/gdrive')
```

```python
# Add the Python module folder to path 
import sys
import os

gDrive_location = '/content/gdrive/MyDrive'
module_folder = 'Utilities4Colab'

full_path_module_folder = os.path.join(gDrive_location, module_folder)
sys.path.append(full_path_module_folder)

# Import class and define the instance as cv (therby you can delete only '2' below)
from cv2workaround import Cv2Workaround 
cv = Cv2Workaround()
```

```python
import cv2

# Load our input image
path_image = os.path.join(gDrive_location, 'test.jpg')
image = cv2.imread(path_image)
assert image is not None, 'image not found'


# cv2.imshow('Beach', image) # Original line
cv.imshow('Beach', image) # just "2"  is removed

cv2.waitKey()
cv2.destroyAllWindows()
```
